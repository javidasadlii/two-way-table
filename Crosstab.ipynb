```python
import pandas as pd
df = pd.DataFrame({'Grade': ['A','A','A','B','B', 'B', 'B', 'C', 'D', 'D'],
                   'Age': [18, 18, 18, 19, 19, 20, 18, 18, 19, 19],
                   'Gender': ['M','M', 'F', 'F', 'F', 'M', 'M', 'F', 'M', 'F']})
df
```

```python
pd.crosstab(index= df["Age"], columns = "count")
```

```python
tab = pd.crosstab(index = df["Gender"], columns = df["Age"])
tab
```

```python
tab.index = ["Female","Male"]
tab.columns = [18,19,20]
tab
```

```python
tab1 = pd.crosstab(index = df["Gender"], columns = df["Age"],
                  margins = True)
tab1.index = ["Female","Male", "Coltotal"]
tab1.columns = [18,19,20,"Rowtotal"]
tab1
```
